### << 자연어 대화 인터페이스 >>
* 자연어 대화 인터페이스에는 '언어이해' , '대화관리' , '응답생성' 의 과정이 필요하다
 1. **언어이해** : 자연어 처리 기본 기술
 2. **대화관리** : 목적기반 대화모델 많이 사용(Goal Oriented Dialog Model)
> * 사용하는 추론 방법에 따라 **'규칙 기반 대화 모델'**과 **'통계 기반 대화 모델'**로 나눈다
> * 최근에는 **예제 기반 대화 모델(Example Based Dialog Model)** 등의 통계 기반 방법 많이 연구
> * 대화 예제(dialog example)로부터 언어 분석을 이용해 대화 예제 색인(index)을 구성하고 이로부터 사용자 발화와 가장 유사한 예제를 찾아 이를 시스템 응답에 이용한다.
>* 또 다른 대화모델로 **강화학습**을 사용하는 **POMDP(Partially Observable Markov Decision Process)기반 대화 모델**이 있다.
> * POMDP 기반 대화 모델은 대화 학습 말뭉치를 마코프 결정 프로세스(MDP)를 통해 강화 학습한 대화 모델을 이용하여 최적의 응답을 생성하는 방법이다.
>* 최근에는 알파고에서 사용된 **딥러닝 기반 강화학습**인 **DQN(Deep Q-Network)**이 개방되면서 이를 이용한 연구가 많이 진행되고 있다.

##### **[ 카메라 + 딥러닝 + 강화학습을 활용한 자율주행 ]**
* 고가의 특화센서와 자동차 분야의 전문가가 중심이 되었던 자율주행 기술을 인공지능 분야의 전문가들이 저가의 범용센서를 활용해 자율주행 구현 가능
1. 시각인식 지능의 적용 : 인간의 수준을 넘음(인간_94.90% , MS_97.85%_2017년 기준)
>- Nvidia 인간의 시각으로 인식하기 어려운 물체들까지 인공지능이 높은 성능으로 인식
>- AutoX 센서에 의존하지 않고, 카메라를 통해 입력된 영상 정보만으로 자율주행 구현
2. 학습 지능의 적용 : 사람이 운전을 배워가는 과정과 유사
>- drive.ai 카메라를 통한 인식 과정부터 주행 기능 구현에 이르기까지 자율주행 전 과정 딥러닝만으로 구현
> (차량 간 거리조정, 출동 방지, 표지판, 신호등 인식, 완전 자율주행 가능 수준까지 기능 구현_비가 오거나 밤과 같은 상황의 데이터도 학습 시키며 고도화 중)
3. 강화학습 적용 : 반복학습을 통해 다양한 주행 _ 상황별 대응 방법 터득

##### **[ 사고하는 지능의 적용 ]**
1. 딥마인드가 발표한 인공지능 구현에 관한 논문
* 주변 상황을 개별 정보로 인식하고 이해하는 것을 넘어 각 정보들 사이의 상대적 관계를 논리적으로 파악
>1.  예) 차선 변경시 전방 차량들과의 거리를 개별적으로 인식 → 전방 차량과 거리가 가장 먼 차선으로 변경
>2.  예) 전방 차량의 과속/서행/정속. 급차선변경/차선 이탈 가능성/정상 주행 예측 → 주위 차량을 가급적 회피하며 차량 주행 지속

2. 지능주행 이식
* 인공지능이 새로운 영억에 활용될 때, 적용 분야가 다르더라도 기존 지식을 최대환 활용하기 때문에 단시간에 성능 발휘 가능
* 폭스바겐 자율 주행 기능 구현에 대한 지능의 이식 관련 연구 적용 고려 중
>- 자동차 주행 방식은 공용화
>- 모든 나라에 적용 가능한 주행은 범용적 지능 구현
>- 국가별 차이가 있는 주행 방식은 개별적 재학습 통해 맞춤화
